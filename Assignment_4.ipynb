```python
import requests
import csv
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd

url = "http://www.zongheng.com/rank/details.html?rt=1&d=1"
html = urlopen(url)
soup = BeautifulSoup(html,'html.parser')

books = []
category = []
latest = []
for item in soup.findAll("div", attrs={"class":"rank_d_book_intro fl"}):     
    name = item.find("div",attrs={"class":"rank_d_b_name"})
    cate = item.find("div",attrs={"class":"rank_d_b_cate"})
    last = item.find("div",attrs={"class":"rank_d_b_last"})     
    books.append(name.text)
    category.append(cate.text)
    latest.append(last.text)

df=pd.DataFrame({"Book Name":books,"Author & Category":category, "Latest Chapter":latest})           
df  
df.to_csv("Assignment_4.csv", mode='w', index=False)
```
